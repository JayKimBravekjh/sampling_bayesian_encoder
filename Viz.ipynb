## Hyperparameter studies
In this section we generate plots for the hyperparameter studies.
During these studies we start with the best model and vary hyperparameters to see how the model performance depends on these hyperparameters.

```python
#Load all results
with open('studies/study_mapper.json', 'r') as f:
    study_mapper = json.load(f)
with open('studies/study_mapper_h.json', 'r') as f:
    study_mapper_h = json.load(f)
with open('studies/study_n_draws.json', 'r') as f:
    study_n_draws = json.load(f)        
with open('studies/study_n_draws_h.json', 'r') as f:
    study_n_draws_h = json.load(f)
with open('studies/study_prior_samples_ratio.json', 'r') as f:
    study_prior_samples_ratio = json.load(f)
with open('studies/study_prior_samples_ratio_h.json', 'r') as f:
    study_prior_samples_ratio_h = json.load(f)    
```

```python
def plot_results(results, xlabel, ylabel, filename=None, transform=None):
    """
    Plot cross-validation results
    results: cross-validation results
    xlabel: x label
    ylabel: y label
    filename: file name where the plot will be saved
    transform: a function applied to X axis
    """
    param_values = [val[list(val.keys())[0]] for val in results['params']]
    if transform:
        param_values = [transform(val) for val in param_values]
    plt.plot(param_values, results['mean_test_score'], 'b-o')
    lower = [m - s for m,s in zip(results['mean_test_score'], results['std_test_score'])]
    upper = [m + s for m,s in zip(results['mean_test_score'], results['std_test_score'])]
    plt.fill_between(param_values, lower, upper, color='lightgrey')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    if filename:
        plt.savefig(filename)
    plt.show();
```

```python
plot_results(study_mapper, 'Mapper', 'Accuracy', filename='mapper_c.png');
```

```python
plot_results(study_n_draws, 'Number of Draws', 'Accuracy', filename='draws_c.png');
```

```python
from math import log10
plot_results(study_prior_samples_ratio, 'Log10 of prior distribution weight', 'Accuracy', transform=log10, 
             filename='prior_c.png');
```

```python
plot_results(study_mapper_h, 'Mapper', 'Accuracy', filename='mapper_h.png');
```

```python
plot_results(study_n_draws_h, 'Number of Draws', 'Accuracy', filename='draws_h.png');
```

```python
plot_results(study_prior_samples_ratio_h, 'Log10 of prior distribution weight', 'Accuracy', transform=log10, 
             filename='prior_h.png');
```
